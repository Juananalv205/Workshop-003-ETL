```python
# Function to find duplicate countries within a DataFrame
def find_duplicate_countries(data, year, country_column):
    # Count the occurrences of each country
    country_counts = data[country_column].value_counts()  
    # Filter the countries that appear more than once
    duplicate_countries = country_counts[country_counts > 1]
    # Print the results
    if not duplicate_countries.empty:
        print(f"Duplicates in {year}:")
        print(duplicate_countries)
    else:
        print(f"There are no duplicate countries in {year}.")

# Check for duplicates for each year
find_duplicate_countries(data_2015, "2015", "Country")
find_duplicate_countries(data_2016, "2016", "Country")
find_duplicate_countries(data_2017, "2017", "Country")
find_duplicate_countries(data_2018, "2018", "Country or region")    
find_duplicate_countries(data_2019, "2019", "Country or region")
```

```python
# Extract unique countries from each dataset
countries_2015 = set(data_2015['Country'])
countries_2016 = set(data_2016['Country'])
countries_2017 = set(data_2017['Country'])
countries_2018 = set(data_2018['Country or region'])
countries_2019 = set(data_2019['Country or region'])

# Compare the sets of countries
all_countries = countries_2015 & countries_2016 & countries_2017 & countries_2018 & countries_2019  # Intersection of all sets
country_counts = {
    "2015": len(countries_2015),
    "2016": len(countries_2016),
    "2017": len(countries_2017),
    "2018": len(countries_2018),
    "2019": len(countries_2019),
    "Common Countries": len(all_countries)
}

# Display the results
print(country_counts)

# Check if all datasets have the same number of countries
if all(len(countries) == len(countries_2015) for countries in [countries_2016, countries_2017, countries_2018, countries_2019]):
    print("All datasets have the same number of countries.")
else:
    print("The datasets have different numbers of countries.")
```

```python
# Create sets of (country, region) pairs for 2015 and 2016
countries_regions_2015 = dict(zip(data_2015['Country'], data_2015['Region']))  # Change 'Region' to the correct column name if necessary
countries_regions_2016 = dict(zip(data_2016['Country'], data_2016['Region']))  # Change 'Region' to the correct column name if necessary

# Create a list to store the differences
differences_list = []

# Iterate over the countries from 2015
for country in countries_regions_2015:
    if country in countries_regions_2016:  # Only consider countries that are in both years
        region_2015 = countries_regions_2015[country]
        region_2016 = countries_regions_2016[country]
        
        if region_2015 != region_2016:  # Only add if there is a difference in the region
            differences_list.append({
                'Country': country,
                'Region Year 2015': region_2015,
                'Region Year 2016': region_2016
            })

# Create a DataFrame from the list of differences
differences_df = pd.DataFrame(differences_list)

# Count the differences
num_differences = len(differences_df)

# Print results
if num_differences == 0:
    print("There are no differences in regions for common countries in 2015 and 2016.")
else:
    print(f"There are {num_differences} different (Country, Region) pairs between 2015 and 2016 for common countries.")
    differences_df  # Display the DataFrame with the differences
    
```

```python
# Diccionario de renombramiento
rename_dict = {
    'Country': 'country',
    'Country or region': 'country',
    'Happiness Rank': 'happiness_rank',
    'Happiness.Rank': 'happiness_rank',
    'Overall rank': 'happiness_rank',
    'Happiness Score': 'happiness_score',
    'Happiness.Score': 'happiness_score',
    'Score': 'happiness_score',
    'Freedom': 'freedom',
    'Freedom to make life choices': 'freedom',
    'Economy (GDP per Capita)': 'economy_GDP_per_capita',
    'Economy..GDP.per.Capita.': 'economy_GDP_per_capita',
    'GDP per capita': 'economy_GDP_per_capita',
    'Social support': 'family',
    'Family': 'family',
    'Health (Life Expectancy)': 'health_life_expectancy',
    'Health..Life.Expectancy.': 'health_life_expectancy',
    'Healthy life expectancy': 'health_life_expectancy',
    'Trust (Government Corruption)': 'government_corruption',
    'Trust..Government.Corruption.': 'government_corruption',
    'Perceptions of corruption': 'government_corruption',
    'Generosity': 'generosity',
    'Region': 'region'
}

# Crear un conjunto de las columnas est√°ndar (valores √∫nicos del diccionario) e incluir 'region' en todos los a√±os
standard_columns = set(rename_dict.values()).union({'region'})

# Renombrar las columnas y asegurar que 'year' est√© al final en cada DataFrame
for df, year in zip([data_2015, data_2016, data_2017, data_2018, data_2019], [2015, 2016, 2017, 2018, 2019]):
    # Renombrar columnas
    df.rename(columns=rename_dict, inplace=True)
    
    # A√±adir la columna 'region' con valores NaN si no existe en el DataFrame actual
    if 'region' not in df.columns:
        df['region'] = np.nan
    
    # Filtrar para conservar solo las columnas est√°ndar, incluida 'region'
    df = df.loc[:, [col for col in df.columns if col in standard_columns]]
    
    # A√±adir la columna 'year'
    df['year'] = year
    
    # Reordenar las columnas para que 'year' est√© al final
    cols = [col for col in df.columns if col not in ['year']] + ['year']
    df = df.loc[:, cols]  # Esto reordenar√° las columnas y pondr√° 'year' al final

    # Actualizar el DataFrame original
    if year == 2015:
        data_2015 = df
    elif year == 2016:
        data_2016 = df
    elif year == 2017:
        data_2017 = df
    elif year == 2018:
        data_2018 = df
    elif year == 2019:
        data_2019 = df
```

```python
# Paso 1: Filtrar los a√±os 2015 y 2016 para crear el diccionario de mapeo de regiones
# Filtrar para obtener solo las filas de 2015 y 2016, y seleccionar las columnas relevantes
region_mapping_df = df_concat[df_concat['year'].isin([2015, 2016])][['country', 'region']]

# Eliminar duplicados para asegurarnos de que cada pa√≠s tiene una regi√≥n √∫nica
region_mapping_df.drop_duplicates(subset=['country', 'region'], inplace=True)

# Crear el diccionario de mapeo donde la clave es el pa√≠s y el valor es la regi√≥n
region_mapping = region_mapping_df.set_index('country')['region'].to_dict()

# Paso 2: Rellenar las regiones faltantes en df_concat utilizando el diccionario
# Usamos map para asignar las regiones de acuerdo con el diccionario, sin sobreescribir las existentes
df_concat['region'] = df_concat['region'].fillna(df_concat['country'].map(region_mapping))

# Ahora `df_concat` tiene las regiones faltantes rellenadas usando los datos de 2015 y 2016.
```

```python
print(f'In this moment have {df_concat['region'].isnull().sum()} countrys without region')
```

```python
# Filtrar los pa√≠ses donde la columna 'region' sigue siendo NaN
paises_sin_region = df_concat[df_concat['region'].isna()]['country'].unique()

# Mostrar la lista de pa√≠ses sin regi√≥n
print("Countries without an assigned region after the clean-up process:")
print(paises_sin_region)
```

```python
# Diccionario para asignar manualmente las regiones de los pa√≠ses faltantes
regiones_manual = {
    'Taiwan Province of China': 'Eastern Asia',
    'Hong Kong S.A.R., China': 'Eastern Asia',
    'Trinidad & Tobago': 'Latin America and Caribbean',
    'Northern Cyprus': 'Middle East and Northern Africa',
    'North Macedonia': 'Central and Eastern Europe',
    'Gambia': 'Sub-Saharan Africa'
}

# Asignar las regiones a los pa√≠ses faltantes
df_concat['region'] = df_concat.apply(
    lambda row: regiones_manual.get(row['country'], row['region']),
    axis=1
)

# Verificar si hay pa√≠ses que todav√≠a no tienen regi√≥n asignada
paises_sin_region_actualizado = df_concat[df_concat['region'].isna()]['country'].unique()
print("Countries with no region assigned after manual update:")
print(paises_sin_region_actualizado)
```
